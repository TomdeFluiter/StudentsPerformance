```python
import pandas as pd
import statsmodels.formula.api as sm
```

```python
Facebook = pd.read_csv('https://raw.githubusercontent.com/TomdeFluiter/StudentsPerformance/refs/heads/main/Facebook.csv')
Facebook.head()
```

```python
#recode
Facebook_dummies = pd.get_dummies(Facebook['Payment'])
Facebook2 = pd.concat([Facebook,Facebook_dummies], axis=1)
Facebook2.head()

model1 = sm.ols('like~Paid+Not_Paid', data=Facebook2).fit()
print(model1.summary())
```
