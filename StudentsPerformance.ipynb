```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as scipy
import statsmodels.formula.api as sm

StudentsPerformance = pd.read_csv("StudentsPerformance.csv")
print(StudentsPerformance.head())

reading_score = StudentsPerformance.reading_score
math_score = StudentsPerformance.math_score

plt.figure()
plt.scatter(reading_score, math_score)

writing_score = StudentsPerformance.writing_score
plt.figure()
plt.scatter(reading_score, writing_score)

plt.show()

scipy.pearsonr(reading_score, math_score)
correlation, pvalue = scipy.stats.pearsonr(reading_score, writing_score)
print(f"Pearson correlation: {correlation}")
print(f"P-Vaulue: {pvalue}")

numeric_df = StudentsPerformance.select_dtypes(include=["number"])
matrix = numeric_df.corr()
print (matrix)

model1 = sm.ols("reading_score ~ writing_score", data = StudentsPerformance).fit()
print(model1.summary())
```
