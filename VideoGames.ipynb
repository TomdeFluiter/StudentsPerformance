```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as scipy
import statsmodels.formula.api as sm
import seaborn as sn
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif 
from statsmodels.tools.tools import add_constant

VideoGames = pd.read_csv("VideoGames.csv")
print(VideoGames.head())

model1 = sm.ols ('VideoGames["US Sales (millions)"] ~ Platform', data = VideoGames).fit()
print(model1.summary())

model2 = sm.ols ('VideoGames["US Sales (millions)"] ~ Lifecycle', data = VideoGames).fit()
print(model2.summary())

model3 = sm.ols ('VideoGames["US Sales (millions)"] ~ Usedprice', data = VideoGames).fit()
print(model3.summary())

model4 = sm.ols ('VideoGames["US Sales (millions)"] ~ YearReleased', data = VideoGames).fit()
print(model4.summary())

model5 = sm.ols ('VideoGames["US Sales (millions)"] ~ Platform+Lifecycle+Usedprice+YearReleased', data = VideoGames).fit()
print(model5.summary())

X = VideoGames[["Platform","Lifecycle","Usedprice","YearReleased"]].dropna()
X = add_constant(X)

VIF_data = pd.DataFrame()
VIF_data["variable"] = X.columns
VIF_data["VIF"] = [vif(X,i)
                   for i in range(len(X.columns))]
print(VIF_data)
```
