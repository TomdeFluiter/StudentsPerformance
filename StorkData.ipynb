```python
import pandas as pd
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
import seaborn as sn

Stork = pd.read_csv("https://raw.githubusercontent.com/TomdeFluiter/StudentsPerformance/main/Stork.csv")
Stork.head()
```

```python
Model = sm.ols('BirthRate ~ StorkPairs+LandArea', data = Stork).fit()
print(Model.summary())
```

```python
plt.figure()
sn.regplot(x=Stork.LandArea,y=Stork.BirthRate)
sn.regplot(x=Stork.LandArea,y=Stork.StorkPairs)
plt.show()
```
